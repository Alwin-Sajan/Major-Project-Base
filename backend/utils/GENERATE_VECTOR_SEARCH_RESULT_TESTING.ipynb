```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings
import config
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name=config.EMBEDDING_E5_LARGE,   
    model_kwargs={
    "trust_remote_code": True,
    "device": "cuda" 
    },
    encode_kwargs={"normalize_embeddings": True},
)

# Load vectorstore
vectorstore = FAISS.load_local(
    config.DB_FAISS_PATH,
    embeddings,
    allow_dangerous_deserialization=True
)
```

```python
user_input = '''
Which fish species has a deep, compressed oval body with 8 dorsal spines, elongated second dorsal and anal spines, a small downward-pointing mouth with villiform teeth, silvery abdomen with faint grey-brown streaks and yellow flank blotches, and is found in Palk Bay, Kakinada, and Cochin?
'''

docs_and_scores = vectorstore.similarity_search_with_score(user_input, k=5)
context = "\n\n".join(
    f"[Doc {i+1}]\n{d[0].page_content}"
    for i, d in enumerate(docs_and_scores)
)
print(context)
```
