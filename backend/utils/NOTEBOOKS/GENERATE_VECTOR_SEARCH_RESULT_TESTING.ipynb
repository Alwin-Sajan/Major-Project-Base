```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings
import config
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name=config.EMBEDDING_E5_LARGE,   
    model_kwargs={
    "trust_remote_code": True,
    "device": "cuda" 
    },
    encode_kwargs={"normalize_embeddings": True},
)

# Load vectorstore
vectorstore = FAISS.load_local(
    config.DB_FAISS_PATH,
    embeddings,
    allow_dangerous_deserialization=True
)
```

```python
user_input = '''
Which fish species has a deep, compressed oval body with 8 dorsal spines and 16 rays, 3 anal spines and 14 rays, 13 pectoral rays, a deeply forked caudal fin, ventral fins with axillary scales not reaching the anal origin, two supraorbital spines, a finely serrated preopercular margin, a large horizontal mouth with villiform teeth and paired symphysial canines, a pointed snout, silvery-grey coloration with irregular dark marks and circles on the back, black-dotted pectoral base, dusky posterior caudal margin, and is found around Great Nicobar Island and off Cochin in southern India
'''

docs_and_scores = vectorstore.similarity_search_with_score(user_input, k=5)
context = "\n\n".join(
    f"[Doc {i+1}]\n{d[0].page_content}"
    for i, d in enumerate(docs_and_scores)
)
print(context)
```
