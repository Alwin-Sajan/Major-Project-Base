```python
import json
```

```python
#CONFIG
JSONL_PATH = r"/home/abk/abk/projects/Major-project-basic-ui/backend/taxonomy_data/merged_taxonomic_chunks.jsonl"
```

```python
def load_taxonomic_for_game(jsonl_path: str):
    entries = []

    with open(jsonl_path, "r", encoding="utf-8") as f:
        for line in f:
            data = json.loads(line)

            if data.get("chunk_type") != "taxonomic":
                continue

            entries.append({
                "species_name": data.get("species_name"),
                "genus": data.get("genus"),
                "text": data.get("text"),
            })

    return entries
```

```python
print(len(load_taxonomic_for_game(JSONL_PATH)))
```
